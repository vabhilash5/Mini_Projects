```python
dic={}
for i in range(df.shape[0]):
    if df['source_center'][i] in dic:
        continue
    else:
        if (df['source_name'][i]) is not np.nan:
            dic[df['source_center'][i]]=df['source_name'][i]
for i in range(df.shape[0]):
    if df['destination_center'][i] in dic:
        continue
    else:
        if (df['destination_name'][i]) is not np.nan:
            dic[df['destination_center'][i]]=df['destination_name'][i]

for i in range(df.shape[0]):
    if df['source_name'][i] is np.nan:
        if df['source_center'][i] in dic:
            df['source_name'][i]=dic[df['source_center'][i]]

for i in range(df.shape[0]):
    if df['destination_name'][i] is np.nan:
        if df['destination_center'][i] in dic:
            df['destination_name'][i]=dic[df['destination_center'][i]]
```

```python
for i in range(df.shape[0]):
    if df['source_name'].loc[i] is np.nan:
        df['source_name'].loc[i]=df['source_center'].loc[i]
for i in range(df.shape[0]):
    if df['destination_name'].loc[i] is np.nan:
        df['destination_name'].loc[i]=df['destination_center'].loc[i]
```
